#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
data = fitbit[['Minutes Lightly Active', 'Minutes Sedentary', 'Minutes Very Active', 'Steps']]
data


x_col = 'Steps'
y_columns= ['Minutes Lightly Active', 'Minutes Sedentary', 'Minutes Very Active']


for elementy in y_columns:

    figure = plt.figure
    ax = plt.gca()
    ax.scatter(data[x_col], data[elementy])
    ax.set_xlabel(x_col)
    ax.set_ylabel(elementy)
    ax.set_title("{} vs {}".format(x_col, elementy))

    plt.legend()
    plt.show()
    
```

#### Plot the correlation matrix using seaborn.

```python
corr= titanic.corr()
    
mask = np.triu(np.ones_like(corr, dtype=bool)) 
color_map = sns.diverging_palette(0, 10, as_cmap=True)
sns.heatmap(corr,  
            mask = mask,
            cmap=color_map,
            square=True, 
            linewidth=0.5,
            vmax=1, #defining max range
            vmin=-1, #defining min range
            cbar_kws={"shrink": .5} #steps for the scale
);
```

```python
"""
Class and Fare are the most correlated values. 
"""
```
